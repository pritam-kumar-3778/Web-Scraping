```python
from bs4 import BeautifulSoup

# Read the scrapped data
with open("WS_Scrapped_data/data1.html", "r") as f:
    html_content = f.read()

# Beautifulsoup is a class, Now we create a object called soup
soup = BeautifulSoup(html_content, "lxml")
```

```python
# Extract country name and population from the data

all_h3 = soup.find_all("h3")

for h3 in all_h3:
    print(h3)
```

```python
# Extract only the country name
# get_text() : Use for extracting content only
# strip = True : Use for remove all unnecessery spaces

for h3 in all_h3:
    print(h3.get_text(strip=True))
```

```python
# Print Population of each country
# population is under div -> span

for h3 in all_h3:
    print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))
```
